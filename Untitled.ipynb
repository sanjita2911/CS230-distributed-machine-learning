```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# 1) Load and fix timestamps
records = []
with open('metrics_raw.json','r') as f:
    for line in f:
        rec = json.loads(line)
        for ts in ('started_at','finished_at'):
            rec[ts] = rec[ts].rstrip('Z')
        records.append(rec)

df = pd.DataFrame(records)
df['started_at'] = pd.to_datetime(df['started_at'], utc=True)
df['finished_at'] = pd.to_datetime(df['finished_at'], utc=True)

# 2) Compute execution duration in seconds
df['exec_secs'] = (df['finished_at'] - df['started_at']).dt.total_seconds()

# 3) We want a continuous time axis: convert times to matplotlib dates
df['start_num'] = mdates.date2num(df['started_at'])
df['end_num']   = mdates.date2num(df['finished_at'])

# 4) Assign each worker an integer y-position
workers = sorted(df['worker_id'].unique())
worker_to_y = {w: i for i,w in enumerate(workers)}
df['y'] = df['worker_id'].map(worker_to_y)

# 5) Plot: each subtask as a colored horizontal bar from start→end on its worker row
fig, ax = plt.subplots(figsize=(12, 6))

# Choose a colormap for exec_secs (longer = brighter)
cmap = plt.get_cmap('Reds')

for _, row in df.iterrows():
    ax.barh(
        y=row['y'],
        width=(row['end_num'] - row['start_num']),
        left=row['start_num'],
        height=0.8,
        color=cmap(row['exec_secs'] / df['exec_secs'].max()),
        edgecolor='none'
    )

# 6) Formatting
ax.set_yticks(range(len(workers)))
ax.set_yticklabels(workers)
ax.set_xlabel('Time')
ax.set_ylabel('Worker ID')
ax.set_title('Worker Execution Timeline (color = relative exec duration)')

# Format x-axis as HH:MM:SS
ax.xaxis_date()
ax.xaxis.set_major_formatter(mdates.DateFormatter('%H:%M:%S'))

plt.tight_layout()
plt.show()
```
